```python
import pandas as pd
```

```python
import seaborn as sb #visualization için kullanacağım kütüphane
import matplotlib.pylab as plt
```

```python
sb.set(color_codes=True)
```

```python
vG = pd.read_csv('vgsales.csv')
```

```python
vG.info() #dosyanın yapısı ile ilgili bilgiler
```

```python
vG.head(10) #dosyanın içi / dünya üzerinde en çok satan ilk 10 oyun
```

```python
sb.barplot(vG['Genre'], vG['Global_Sales']) #Hangi genre'nın globalde ne kadar satıldığını($million olarak) gösteren chart
plt.xticks(rotation=45) #Tablonun daha rahat anlaşılabilmesi için x-eksenindeki değerler 45 derece döndürüldü
```

```python
sb.barplot(vG['Platform'], vG['Global_Sales']) #Hangi platformda toplam ne kadar oyun satıldığını gösteren chart
plt.xticks(rotation=90)
```

```python
sb.histplot(vG['Genre']) #Genre'ya göre oyun adetini histogram olarak gösteren chart
plt.xticks(rotation=45)
```

```python
sb.jointplot(vG['Genre'], vG['Global_Sales']) #En baştaki chart gibi satışların noktalı grafik olarak gösterimi
```

```python
sb.jointplot(vG['NA_Sales'], vG['Global_Sales']) #Yalnızca Kuzey amerika satışlarının global satışlar içindeki önemini gösteren chart. Neredeyse toplam satışın yarısını oluşturuyor.
```

```python
sb.jointplot(vG['Other_Sales'], vG['Global_Sales']) #Diğer ülkelerdeki satışlar ile global'i karıştıran bir chart. Yorumlayacak olursak bazı oyunlar bazı ülkelerin kültürüne daha yakın oluyor ve oralarda daha fazla kopya satıyor dünya genelinde.
```

```python
sb.jointplot(vG['Year'],vG['EU_Sales']) #Yıllara göre avrupada oyun satışını gösteren bir chart 
```

```python
sb.countplot(vG['Publisher']) #Publisher'ların yayınladığı oyun sayılarını gösteren bir chart
```

```python
sb.barplot(vG['Year'], vG['JP_Sales']) #Yıllara göre Japonya'daki oyun satışını gösteren chart
plt.xticks(rotation=90)
```

```python
sb.pairplot(vG[['Year', 'NA_Sales', 'EU_Sales']])
```

```python
sb.stripplot(vG['Year'], vG['EU_Sales'])
plt.xticks(rotation=90)
```
